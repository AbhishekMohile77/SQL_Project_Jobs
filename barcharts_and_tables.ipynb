# Import Libraries



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# First sample chart for total jobs

```python
df = pd.read_csv('/content/sample query chart.csv')
```

```python
df
```

```python
df_sorted = df.sort_values(by='total_jobs', ascending = False)
df_sorted
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Set black background for the figure and axes
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

plt.barh(df['job_title_short'], df['total_jobs'], color='lightblue', edgecolor ='green')

# Add chart title and labels
plt.title('Total Number of Available Jobs for Top 10 Job Posts in 2023', fontsize=18)
plt.xlabel('Total Jobs', fontsize=14)
plt.ylabel('Job Posts', fontsize=14)

ax.xaxis.label.set_color('silver')
ax.yaxis.label.set_color('silver')
ax.title.set_color('white')
ax.tick_params(colors='silver', which='both')

ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')

plt.grid(axis = 'x', color = 'white')
plt.tight_layout()
plt.show()
```

# Query 2

```python
df2 = pd.read_csv('/content/Q2 top job skills.csv')
df2
```

```python
skills_count = df2.skills.value_counts().head(10)
skills_count
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Set black background for the figure and axes
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

#plt.barh(df1['job_title'].head(10), df1['salary_year_avg'].head(10), color='teal', edgecolor ='green')
skills_count = skills_count.sort_values(ascending=True)
skills_count.plot(kind='barh', color='#872C64')

# Add chart title and labels
plt.title('Skills Required for Top Paying Data Analyst Jobs in 2023', fontsize=18)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Skill', fontsize=14)

ax.xaxis.label.set_color('silver')
ax.yaxis.label.set_color('silver')
ax.title.set_color('white')
ax.tick_params(colors='silver', which='both')

ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')

plt.grid(axis = 'x', color = 'white')
plt.tight_layout()
plt.show()
```
